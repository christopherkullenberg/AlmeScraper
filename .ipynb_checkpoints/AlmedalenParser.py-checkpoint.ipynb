```python
import urllib.request
import re
from bs4 import BeautifulSoup
```

```python
urllist = []
urlfile = open('almeurl.txt')
almeurls = urlfile.readlines()
for a in almeurls:
    urllist.append(a)
```

```python
eventlist = []
pagecounter = 0 
```

```python
for url in urllist:
    with urllib.request.urlopen(url) as event:
        eventdict = {}
        html = event.read()
        soup = BeautifulSoup(html, 'html.parser')
        datadivs = soup.findAll("div", { "class" : "row clearfix" })
        for d in datadivs:
            leftdivs = d.findAll("div", { "class" : "leftcol" })
            rightdivs = d.findAll("div", { "class" : "rightcol" })
            def leftdiv():
                for l in leftdivs:
                    for x in l:
                        return x
            def rightdiv():
                for r in rightdivs:
                    for x in r:
                        return x.strip()
            d['mynewkey'] = 'mynewvalue'
            eventdict[leftdiv()] = rightdiv()
    eventlist.append(eventdict)
    pagecounter += 1
    print("Added to dictionary. Total count: " + str(pagecounter))
```

```python
for e in eventlist:
    print(e)
```
